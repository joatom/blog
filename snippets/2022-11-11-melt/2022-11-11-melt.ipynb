```python
#|echo: false

import pandas as pd

data = (pd.read_csv('../../data/harvest/fruit.csv')
          .merge(pd.read_csv('../../data/harvest/harvest.csv'), on='fruit_id')[['date','name','amount','quality']]
          .sort_values(['date','name'])
          .reset_index(drop=True)
          .rename(columns={'name': "fruit"})
       )

data
```

# Pivot

```python
data_pv = data.pivot(index='date', columns='fruit', 
                     values=['amount','quality']).reset_index()
data_pv
```

# Melt (Unpivot)

```python
data_pv.columns = ['date', 'apple (red)_amt', 'grape (red)_amt',
                   'grape (white)_amt','apple (red)_qlt',
                   'grape (red)_qlt','grape (white)_qlt']
```

```python
data_pv.melt(id_vars='date', value_vars=['apple (red)_amt', 
                                         'grape (red)_amt',
                                         'grape (white)_amt'])
```
